Analise de dados do ENEM 2021

```python
Verificar se há correlação entre a formação acadêmica dos pais com a nota dos participantes
```

```python
import pandas as pd
```

```python
#Importando a base de dados como um dataframe e verificando sua estrutura
basededados = pd.read_csv("MICRODADOS_ENEM_2021.csv",sep=";",encoding="ISO-8859-1")
basededados.head()
```

```python
'''Vamos utilizar as notas de matemática e redação, comparando-as com os dados
socio econômicos Q001('Até que série seu pai, ou homem responsável por você,estudou?')
Q002('Até que série sua mãe, ou a mulher responsável por você,estudou?')'''

selecionadas=['NU_INSCRICAO','NU_NOTA_MT','NU_NOTA_REDACAO','Q001','Q002']
dadosselecionados = basededados.filter(items=selecionadas)
dadosselecionados.head()
```

```python
#Removendo dados NAN
dadosselecionados = dadosselecionados.dropna()
dadosselecionados.head()
```

```python
#Criar um dicionário para ter melhor visualização dos dados no dataframe
dicionario1e2={'A':'Nunca estudou',
               'B':'Não completou a 4ª série/5º ano do Ensino Fundamental.',
              'C':'Completou a 4ª série/5º ano, mas não completou a 8ª série/9º ano do Ensino Fundamental.',
              'D':'Completou a 8ª série/9º ano do Ensino Fundamental, mas não completou o Ensino Médio.',
              'E':'Completou o Ensino Médio, mas não completou a Faculdade.',
              'F':'Completou a Faculdade, mas não completou a Pós-graduação.',
              'G':'Completou a Pós-graduação.',
              'H':'Não sei.'}
```

```python
#Inserindo a coluna NO_Q001 com o dicionário criado
dadosselecionados['NO_Q001'] = [dicionario1e2[resp] for resp in dadosselecionados.Q001]
dadosselecionados.head()
```

```python
#Fazer o mesmo processo para o Q002
dadosselecionados['NO_Q002'] = [dicionario1e2[resp] for resp in dadosselecionados.Q002]
dadosselecionados.head()
```

Distribuição dos candidatos do Enem 2021 por escolaridade do Pai/responsável:

```python
dadosselecionados.filter(items=['NU_INSCRICAO','NO_Q001']).groupby('NO_Q001').count().sort_values(by='NU_INSCRICAO',ascending=False)
```

Distribuição dos candidatos do Enem 2021 por escolaridade da Mãe/responsável:

```python
dadosselecionados.filter(items=['NU_INSCRICAO','NO_Q002']).groupby('NO_Q002').count().sort_values(by='NU_INSCRICAO',ascending=False)
```

Verificando o desempenho em matemática segundo a escolaridade do Pai/responsável:

```python
dadosselecionados.filter(items=['NU_NOTA_MT','NO_Q001']).groupby('NO_Q001').mean().sort_values(by='NU_NOTA_MT',ascending=False)
```

Verificando o desempenho em matemática segundo a escolaridade da Mãe/responsável:

```python
dadosselecionados.filter(items=['NU_NOTA_MT','NO_Q002']).groupby('NO_Q002').mean().sort_values(by='NU_NOTA_MT',ascending=False)
```

Verificando o desempenho da redação segundo a escolaridade do Pai/responsável:

```python
dadosselecionados.filter(items=['NU_NOTA_REDACAO','NO_Q001']).groupby('NO_Q001').mean().sort_values(by='NU_NOTA_REDACAO',ascending=False)
```

Verificando o desempenho da redação segundo a escolaridade da Mãe/responsável:

```python
dadosselecionados.filter(items=['NU_NOTA_REDACAO','NO_Q002']).groupby('NO_Q002').mean().sort_values(by='NU_NOTA_REDACAO',ascending=False)
```

Verificando o desempenho da redação segundo a escolaridade da Mãe/responsável, agora utilizando o método describe:

```python
dadosselecionados.filter(items=['NU_NOTA_REDACAO','NO_Q002']).groupby('NO_Q002').describe()
```

Após o tratamento e a manipulação dos dados podemos observar que a maior média dos participantes, tanto em matemática, como na redação, está também, fortemente correlacionada com a formação dos responsáveis. Por mais que há de se relevar que não é um fator determinante, visto que, há participantes que os responsáveis não tem formação acadêmica e mesmo assim obtiveram excelentes resultados, mas que impacta positivamente e negativamente na maior parte dos casos.
